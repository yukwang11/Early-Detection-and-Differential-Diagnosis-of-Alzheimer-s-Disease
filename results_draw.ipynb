```python
import matplotlib.pyplot as plt
import matplotlib
from contextlib import contextmanager
import pandas as pd
import matplotlib.ticker as ticker
from matplotlib.ticker import FixedLocator
import matplotlib.gridspec as gridspec
from mpl_toolkits.axes_grid1 import Grid



@contextmanager
def toggle_matplotlib():
    # Backup original matplotlib settings
    original_settings = plt.rcParams.copy()
    try:
        # Set font for Chinese and adjust unicode settings
        # plt.rcParams["font.family"] = "Microsoft Yahei"  # Change to a Chinese-supported font
        plt.rcParams["axes.unicode_minus"] = False       # Ensure minus signs display correctly
        yield
    finally:
        # Restore original matplotlib settings
        plt.rcParams.update(original_settings)

```

```python
def plot_facet(df, metrics, models, nrows=4, ncols=4, figsize=(16, 16), background_color=None):
    # Set color map for lines
    colors = matplotlib.colormaps.get_cmap("tab10")
    
    # Create figure and grid layout with tighter spacing between subplots
    fig = plt.figure(figsize=figsize)
    grid = Grid(fig, rect=111, nrows_ncols=(nrows, ncols), axes_pad=0.02, label_mode='L')  # Adjust axes_pad for spacing

    for row, metric in enumerate(metrics):
        for col, model in enumerate(models):
            # Select data for the specific metric and model
            y = df[f"{metric}_{model}"]
            x = df["days_before_diag"]
            
            # Get the appropriate axis
            ax = grid[row * ncols + col]  # Access each subplot using Grid index
            
            # Plot each metric-model combination
            ax.plot(x, y, label=f"{metric}_{model}", marker='o', color=colors(col), markersize=3)
            
            # Plot other models for the same metric with lower opacity for context
            for other_model in models:
                if other_model != model:
                    other_y = df[f"{metric}_{other_model}"]
                    ax.plot(x, other_y, color=colors(models.index(other_model)), alpha=0.27)
            
            # Set background color if specified
            if background_color is not None:
                ax.set_facecolor(background_color)

            # Customize axis: Remove top and right spines
            ax.spines["top"].set_visible(False)
            ax.spines["right"].set_visible(False)
            
            # Set y-axis for the first column with labels on the inside (right side of the tick marks)
            if col == 0:
                ax.set_ylabel(metric.capitalize(), fontsize=16, fontweight="bold")
                ax.yaxis.tick_left()
                ax.set_yticks([0.5, 0.6, 0.7, 0.8, 0.9, 1.0])
                ax.tick_params(axis='y', labelleft=True, labelright=False, direction='in', pad=0)  # Position labels near the tick marks
            else:
                # Remove y-axis completely for columns 2, 3, and 4
                ax.spines["left"].set_visible(False)
                ax.yaxis.set_visible(False)  

            # Set column titles at the top for the first row only
            if row == 0:
                ax.set_title(model, fontsize=16, fontweight="bold")
            
            # Display x-axis for the second row only, with labels above the tick marks
            if row == 1:
                ax.xaxis.tick_bottom()
                ax.set_xticks([-730, -540, -365, -270, -180, -90, -30, 0])
                ax.tick_params(axis='x', labelbottom=True, labeltop=False, direction='inout', pad=5, rotation=45)  # Position labels above the tick marks
            else:
                # Hide x-axis for all rows except the second
                ax.spines["bottom"].set_visible(False)
                ax.xaxis.set_visible(False)

            # Set consistent x-axis ticks for alignment across subplots
            specific_x_ticks = [-730, -540, -365, -270, -180, -90, -30, 0]
            ax.set_xticks(specific_x_ticks)
            ax.xaxis.set_major_locator(FixedLocator(specific_x_ticks))
            ax.xaxis.set_tick_params(labelbottom=False)

            # Set y-axis limits and specific ticks for alignment across subplots
            # specific_y_ticks = []
            specific_y_ticks = [0.5, 0.6, 0.7, 0.8, 0.9, 1.0]
            ax.set_ylim(0.5, 1.0)
            ax.yaxis.set_major_locator(FixedLocator(specific_y_ticks))
            ax.yaxis.set_tick_params(labelbottom=False)

            for tick in ax.get_xticks():
                ax.axvline(x=tick, color='gray', linestyle='--', linewidth=0.5, alpha=0.4)

            for tick in ax.get_yticks():
                ax.axhline(y=tick, color='gray', linestyle='--', linewidth=0.5, alpha=0.4)

    # Shared x-axis label at the bottom
    fig.text(0.5, 0.02, 'Days Before Diagnosis', ha='center', fontsize=16)

    plt.tight_layout(rect=[0.05, 0.05, 1, 1])  # Adjust layout to make space for axis labels
    return fig


```

```python
# Usage example (assuming df is already loaded and cleaned, with columns like "accuracy_LSTM", etc.)
metrics = ["accuracy", "recall", "f1", "roc_auc"]
models = ["LSTM-D", "RF", "DT", "LR"]

with toggle_matplotlib():
    fig = plot_facet(df, metrics, models, nrows=4, ncols=4, figsize=(16, 9), background_color="#f9fff3")
    # "#f6f6f6"
    plt.show()
```

```python
import numpy as np

# Reload the file and set up the data again
file_path = "itemid_to_vector.csv"
df_itemid_to_vector = pd.read_csv(file_path)
itemid_to_vector = {str(key): value for key, value in df_itemid_to_vector.set_index('itemid').T.to_dict('list').items()}
```

```python
pairs = [(["51087", "Length of Urine\nCollection"], ["51108","Urine Volume"]), 
         (["51101", "Total Collection Time"], ["51109", "Urine Volume, Total"]), 
         (["52285", "RBC, CSF"], ["51790", "Glucose, CSF"]), 
         (["51180", "CD4 Cells, Percent"], ["51181", "CD4/CD8 Ratio"])]

def prepare_radar_data(itemid_pair, itemid_to_vector):
    vectors = [itemid_to_vector[itemid[0]] for itemid in itemid_pair]
    labels = [f"D{i+1}" for i in range(len(vectors[0]))]
    num_vars = len(labels)
    angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()
    angles += angles[:1]  
    vectors = [vector + [vector[0]] for vector in vectors]  
    return labels, angles, vectors

fig, axes = plt.subplots(1, 4, figsize=(20, 7), subplot_kw=dict(polar=True))
# fig.suptitle("Radar Charts of ItemID Pairs", fontsize=16)

for ax, (itemid1, itemid2) in zip(axes, pairs):
    labels, angles, vectors = prepare_radar_data((itemid1, itemid2), itemid_to_vector)
    for vector, itemid in zip(vectors, [itemid1, itemid2]):
        ax.plot(angles, vector, label=f"{itemid[0]} {itemid[1]}")
        ax.fill(angles, vector, alpha=0.2)
    ax.set_yticklabels([])
    ax.set_xticks(angles[:-1])
    # ax.set_xticklabels(labels, fontsize=15)
    ax.set_xticklabels([])
    # ax.set_title(f"{itemid1[0]} & {itemid2[0]}", fontsize=16)

    ax.legend(loc='lower center', bbox_to_anchor=(0.5, -0.27), prop={'size': 18})

plt.subplots_adjust(wspace=0.001)
plt.tight_layout()
plt.show()
```
